## Customer Segmetation Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ifood_df.csv")
```

```python
df['Total_spending']=df[['MntWines','MntMeatProducts','MntFruits','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
```

```python
df.head
```

```python
df['Total_purchase']=df[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum(axis=1)
```

```python
df.head
```

```python
df[['Income','Total_spending']].sort_values(by='Total_spending',ascending=False).head(100)
```

```python
df[['Income','Total_spending']].corr()
```

```python
X= df[['Income','Total_spending','Total_purchase']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans 
kmeans=KMeans(n_clusters=4,random_state=42)
df['Cluster']=kmeans.fit_predict(X_scaled)
```

```python
df.groupby('Cluster').mean()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(df['Income'],df['Total_spending'], c=df['Cluster'])
plt.title('Customer Segmentation')
plt.xlabel('Customer Income')
plt.ylabel('Total Spending')
plt.show()
```
